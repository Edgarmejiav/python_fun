```python
from gtts import gTTS
from IPython.display import Audio
```

```python
# Texto a convertir a voz
texto = "gTTS es una biblioteca de Python para interactuar con la API de texto a voz de Google Translate."

# Crear objeto gTTS y generar archivo de audio
tts = gTTS(text=texto, lang="es")
tts.save("salida.mp3")

# Reproducir el audio
Audio("salida.mp3")
```

```python
import pyttsx3

# Inicializar el motor de síntesis de voz
engine = pyttsx3.init()

# Establecer el idioma español
engine.setProperty('voice', 'spanish')

# Texto a convertir a voz
texto = 'Hola, ¿cómo puedo ayudarte?'

# Decir el texto
engine.say(texto)

# Guardar la salida de voz en un archivo WAV temporal
engine.save_to_file(texto, 'output.wav')
engine.runAndWait()

```

```python
from IPython.display import Audio
Audio('output.wav')
```
